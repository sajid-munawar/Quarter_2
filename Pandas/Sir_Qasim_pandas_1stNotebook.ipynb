# Map

```python
d1 = {0: "Not Survived",
     1: "Survived"}

df['label1'] = df.Survived.map(d1)
df.head()
```

```python
def abc(x):
    if x==0:
        return "Not Survived"
    else:
        return "Survived"
    
df['label2'] = df.Survived.apply(abc) # you can apply any kind of function on column/columns
```
